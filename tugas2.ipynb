```python
import pandas as pd
import numpy as np

dataset = pd.read_csv("titanic.csv",",")

print(dataset)
```

```python
data_array = np.array(dataset)
dimensi = data_array.shape
print("Jumlah Baris = ",dimensi[0])
print("Jumlah Kolom = ",dimensi[1])
```

```python
data = dataset[['Age','Fare']]
data.head(20)
```

```python
data['Age'] = data['Age'].fillna(dataset.groupby(['Survived'])['Age'].transform('mean'))
data.head(20)
```

```python
newmax = 1
newmin = 0
data_minmax = data[['Age','Fare']]
data_minmax['Age'] = (data['Age'] - data['Age'].min()) * (newmax - newmin) / (data['Age'].max()-data['Age'].min())+newmin
data_minmax['Fare'] = (data['Fare'] - data['Fare'].min()) * (newmax - newmin) / (data['Fare'].max()-data['Fare'].min())+newmin
data_minmax.head(20)
```

```python
data_zscore = data[['Age','Fare']]
data_zscore['Age'] = (data['Age'] - data['Age'].mean()) / data['Age'].std()
data_zscore['Fare'] = (data['Fare'] - data['Fare'].mean())/ data['Fare'].std()
data_zscore.head(20)
```

```python
data_sigmoidal = data[['Age','Fare']]
data_sigmoidal['Age'] = (1 - np.exp(-1*data_zscore['Age']))/(1 + np.exp(-1*data_zscore['Age']))
data_sigmoidal['Fare'] = (1 - np.exp(-1*data_zscore['Fare']))/(1 + np.exp(-1*data_zscore['Fare']))
data_sigmoidal.head(20)
```
